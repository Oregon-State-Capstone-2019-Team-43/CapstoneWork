```python
#This class store the internsity and pitch 
class Pause_Info:
  def __init__(self, intensity, pitch):
    self.intensity = intensity
    self.pitch = pitch

    
#This function extract intensity and pitch info from txt file and put all info in to a dictionary
def readTxT(path):
    f = open(path, "r")
    line = f.readline()
    
    #store the joke title
    Title = line.split('/')[-3]
    Title=Title.replace("\x00","")
    pause_dict = {}

    content=f.read();
    f.close()
    
    #replace all these werid \x00 in python
    #also replace all the extra newline char
    content=content.replace("\x00","")
    content=content.replace("\n\n","\n")
    lines=content.split('\n')
    
    #remove the first and last extra new line char
    lines.pop(0)
    lines.pop(-1)

    count=0
    intensity=0
    pitch=0
    for line in lines:
        if(count%3==0):
            name=line.split('.')[0]
            name=name.split('_')[-1]
            id=int(name)
        elif(count%3==1):
            intensity=round(float(line))
        else:
            #if the pitch is undefined, I will ignore it for later clustering.
            if(line=='--undefined--'):
                count+=1
                continue
            pitch=round(float(line))
            pause_dict[id]=Pause_Info(intensity,pitch)
        count+=1
    return (pause_dict,Title)
```
